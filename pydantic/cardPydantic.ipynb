```python
from pydantic import BaseModel, ValidationError, validator, EmailStr, Field

class cadastroUsuario(BaseModel):
  nome: str
  email: EmailStr
  cidade: str = Field(min_length=3, max_length=20)
  estado: str = Field(min_length=2, max_length=2)
  numResidencia: int
```

```python
def validadorDeCadastro(dados):
  try:
    usuario = cadastroUsuario(**dados)
    print("\n Cadastro validado com sucesso!")
    print(f"Nome: {usuario.nome}")
    print(f"Email: {usuario.email}")
    print(f"Cidade: {usuario.cidade}")
    print(f"Estado: {usuario.estado}")
    print(f"Número da Residência: {usuario.numResidencia}")
  except ValidationError as e:
    print("\n Erro de validação. Os seguintes dados não são válidos: ")
    for erro in e.errors():
      print(f" - Campo: {erro['loc'][0]}")
      print(f" - Mensagem: {erro['msg']}")
    return None
```

```python
#Exemplo com dados corretos

dados_caminho_feliz = {
    "nome": "Livia Oliveira",
    "email": "liviafernanda@discente.ufg.br",
    "cidade": "Goiânia",
    "estado": "GO",
    "numResidencia": 123
}

validadorDeCadastro(dados_caminho_feliz)
```

```python
#Exemplo com erro corrigível de int

dados_caminho_alternativo = {
    "nome": "Livia Oliveira",
    "email": "liviafernanda@discente.ufg.br",
    "cidade": "Goiânia",
    "estado": "GO",
    "numResidencia": "123"
}

validadorDeCadastro(dados_caminho_alternativo)
```

```python
#Exemplo com erro de int

dados_caminho_alternativo1 = {
    "nome": "Livia Oliveira",
    "email": "liviafernanda@discente.ufg.br",
    "cidade": "Goiânia",
    "estado": "GO",
    "numResidencia": "abc"
}

validadorDeCadastro(dados_caminho_alternativo1)
```

```python
#Exemplo com email inválido (EmailStr)

dados_caminho_alternativo2 = {
    "nome": "Lívia Oliveira",
    "email": "liviafernanda",
    "cidade": "Goiânia",
    "estado": "GO",
    "numResidencia": 123
}

validadorDeCadastro(dados_caminho_alternativo2)
```

```python
#serialização
import json

usuario = cadastroUsuario(**dados_caminho_feliz)
print(f"Tipo: {type(usuario)}")
print(f"Conteudo do cadastro: {usuario}")
```
